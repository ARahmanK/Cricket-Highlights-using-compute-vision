```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2, os
```

```python
cap = cv2.VideoCapture('/home/ark/FYP/west.mp4') #capture video from given path
i=0 #intilize variable frame name
fps=cap.get(5) # get the fps of frame
while cap.isOpened():
    ret, frame = cap.read()# Reading video 
    # ret return bool 
    # frame return one frame of video
    if ret== False:
        print('ret is false')
        break;
    current_frame=cap.get(1) # return integer value and tell current number of frame of video
    if current_frame%13==0: # save 2 frame of each sec
        print('frame # ',i)
        path = '/home/ark/FYP/group_meeting'
        cv2.imwrite(os.path.join(path , 'Frame'+str(i)+'.jpg'), frame)
        i += 1
cap.release()
cv2.destroyAllWindows()
```

```python
for rec in range(0,i-1):####insert i as parameter in range 
    img='/home/ark/FYP/group_meeting/Frame'+str(rec)+'.jpg'
    results = model(img)
#     results.show()
    for rec in results.pandas().xyxy:
        if not rec.empty:
            
            if rec.name[0]=='out':
                frame_no=img.split('/home/ark/FYP/group_meeting/Frame')[1][:-4]#[1] list index,[:-4] removing .jpg 
                if frame_no in out_frames:
                    print('OOHH! OUT',frame_no)
                    out.append(frame_no)
                    mix.append(frame_no)
            if rec.name[0]=='four':
                frame_no=img.split('/home/ark/FYP/group_meeting/Frame')[1][:-4]#[1] list index,[:-4] removing .jpg 
                if frame_no in four_frames:
                    print('Hurrah! FOUR ',frame_no)
                    four.append(frame_no)
                    mix.append(frame_no)
            if rec.name[0]=='six':
                frame_no=img.split('/home/ark/FYP/group_meeting/Frame')[1][:-4]#[1] list index,[:-4] removing .jpg 
                print('BigOne! SIX ',frame_no)
                six.append(frame_no)
                mix.append(frame_no)
            
```

```python
from moviepy.editor import *
 
# loading video
complete_video = VideoFileClip('/home/ark/FYP/west.mp4')
 
# clipping of the video
# if out is checked
for rec in out:
    event_sec=int(rec)/2
    clip = complete_video.subclip(int(event_sec), int(event_sec)+15)
    clip.write_videofile(rec+'.mp4')
# if four is checked
for rec in four:
    event_sec=int(rec)/2
    if int(rec)<1000:
        start_lapse=int(event_sec)_5
        end_lapse= int(event_sec)+25
    elif int(rec) < 2300 and int(rec) > 1000:
        start_lapse=int(event_sec)+25  
        end_lapse= int(event_sec)+55
    elif int(rec)<2500 and int(rec)>2300:
        start_lapse=int(event_sec)-30
        end_lapse= int(event_sec)
    else:
        start_lapse=int(event_sec)+30
        end_lapse= int(event_sec)+60
    clip = complete_video.subclip(start_lapse, end_lapse)
    clip.write_videofile(rec+'.mp4')

```
