```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

# Import Model

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5l')
```

# Make Detection with Images

```python
imgs = ['https://i.dawn.com/primary/2021/10/6175908344ee5.jpg']
results = model(imgs)
i=0
for rec in results.pandas().xyxy[0].name:
#     print(str(i+1)+rec)
    if rec=='sports ball':
        print(rec)
    else:
        print('not')
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
results.render()
```

# Real Time Detection

```python
cap = cv2.VideoCapture('/home/ark/Videos/one.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.resize(frame, (800, 400))
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
