```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2, os
```

# READ VIDEO AND CONVERT INTO FRAMES

```python
file_path='/home/ark/FYP/west.mp4'
#capture video from given path
cap = cv2.VideoCapture(file_path) 
#intilize variable frame name
i=0 
# get the fps of frame
fps=cap.get(5)

while cap.isOpened():
    # Reading video 
    # ret return bool 
    # frame return one frame of video
    ret, frame = cap.read()
    #frame ends ret return False
    if ret== False:
        break;
    # return integer value and tell current number of frame of video
    current_frame=cap.get(1)
    # save 2 frame of each sec
    if current_frame%(fps/2)<=1:
        print('frame # ',i)
        path = '/home/ark/FYP/Frames'
        cv2.imwrite(os.path.join(path , 'Frame'+str(i)+'.jpg'), frame)
        i += 1
cap.release()
cv2.destroyAllWindows()
```
